```python
import math
import matplotlib.pyplot as plt
import numpy as np
xl=.4
xu=.48
xr=(xl+xu)/2
def f(x):
    f=math.tan(math.pi*x)-x-6
    return f
actual=0.451047
error=abs(xr-actual)
iteration_number=1
while error > 10**-5:
    if f(xr)*f(xl)<0:
        xu=xr
        xr=(xl+xu)/2
    if f(xr)*f(xl)>0:
        xl=xr
        xr=(xl+xu)/2
    error=abs(xr-actual)
    print(iteration_number)
    iteration_number=iteration_number+1
print(xr)
x_axis=np.linspace(0,1,100)
y_axis=[]
for i in range(len(x_axis)):
    y_axis.append(f(x_axis[i]))
plt.plot(x_axis,y_axis)
plt.grid()
```

Using the interval .4-.48 avoids the discontinuity at 0.5, which would cause a malfunction in the algorithm.
